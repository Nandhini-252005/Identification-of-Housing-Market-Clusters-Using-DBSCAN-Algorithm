```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df.drop('MedHouseVal',axis=1))
scaled_features=scaler.transform(df.drop('MedHouseVal',axis=1))
```

```python
import pandas as pd
x = pd.DataFrame(scaled_features, columns=df.drop('MedHouseVal', axis=1).columns)
# Note: The original intent of dropping 'Genre', 'CustomerID', 'Age' suggests a different dataset was intended.
# If you proceed with the housing dataset, subsequent plotting code expecting 'Annual_Income_(k$)' and 'Spending_Score' will need adjustment.
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=7,min_samples=6,metric='euclidean')
model=db.fit(x)
```

```python
label=model.labels_
label
```

```python
import matplotlib.pyplot as plt
plt.scatter(x['Longitude'], x['Latitude'], c=label, cmap='rainbow')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('DBSCAN Clustering of Housing Data (Longitude vs Latitude)')
plt.show()
```
